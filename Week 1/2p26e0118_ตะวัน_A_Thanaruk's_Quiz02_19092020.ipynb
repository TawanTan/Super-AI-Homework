2. เขียน depth-first search พร้อม อะนิเมชั่น

```
import numpy as np

def operate_dfs(nexts,path,path_ord,node,olders):
  path = path + [node]
  path_ord.append(path)
  n = nexts.get(node)
  if n != None:
    for nextNode in nexts[node]:
      if nextNode != olders:
        operate_dfs(nexts,path,path_ord,nextNode,node)

def cre_path(edges,start):
  nodes = set()
  nexts = {}
  for i in edges:
    nodes.add(i[0])
    nodes.add(i[1])
    if nexts.get(i[0]) == None:
      nexts[i[0]] = []
    if nexts.get(i[1]) == None:
      nexts[i[1]] = []
    nexts[i[0]].append(i[1])
    nexts[i[1]].append(i[0])
  nodes = list(nodes)
  path_ord = []
  operate_dfs(nexts,[],path_ord,start,-1)
  return nodes,nexts,path_ord

def sub_edges(edges1,edges2):
  return [i for i in edges1 if i not in edges2]
```

```
edges = [[1,2],[1,6],[2,5],[2,3],[3,4],[6,7],[7,11],[7,8],
        [8,9],[9,10],[11,12],[3,4],[6,13],[13,14],[1,18],
        [18,21],[21,22],[18,19],[19,20],[2,15],[15,16],[16,17]]
nodes,nexts,path_ord = cre_path(edges,1)
path_ord.append([])
print(nodes)
print(nexts)
path_ord
```
