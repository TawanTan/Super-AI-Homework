```
import copy

#ระบุตำแหน่งของเลขนั้นๆ
def find_loc(rows, element_to_find):
    for i, row in enumerate(rows):
        for j, element in enumerate(row):
            if element == element_to_find:
                return i, j

#ตำแหน่งที่จะสลับกับเลข0ได้
def gen_nodes(state):
    nodes = []
    i, j = find_loc(state, 0)
    for a in [[-1, 0], [+1, 0], [0, -1], [0, +1]]: #ไม่ให้เกินขอบ
        if not ((i+a[0] > 2) or (i+a[0] < 0) or (j+a[1] > 2) or (j+a[1] < 0)):
            nodes.append([i+a[0], j+a[1]])
    return nodes

#คำนวนค่า f-score
def cal_fscore(state, goal):
    count = 0
    for i in range(len(state)):
        for j in range(len(initial[i])):
            if state[i][j] == 0:
                continue
            if state[i][j] != goal[i][j]:
                a, b = find_loc(goal, state[i][j])
                count += abs(i-a) + abs(j-b) #คำนวณค่า g-score
                count += 1 #คำนวณค่า h-score
    return count

#สลับตำแหน่งกับเลข0
def move(states, des):  #des คือตำแหน่งที่จะเคลื่อนไป
    state = copy.deepcopy(states)
    i, j = find_loc(state, 0)
    state[i][j],state[des[0]][des[1]] = state[des[0]][des[1]],state[i][j]
    return state

#astar
def operate(states, goal):
    final = []
    state = copy.deepcopy(states)
    loc0 = find_loc(state, 0)
    while not finish(state, goal):
        nodes = gen_nodes(state) #list ของตำแหน่งที่เลข0เคลื่อนไปได้
        nodes = [nodes[i] for i in range(len(nodes)) if list(nodes[i]) != list(loc0)]
        Min = move(state, nodes[0])
        for i, j in enumerate(nodes):
            if cal_fscore(move(state, j), goal) < cal_fscore(Min, goal):
                Min = move(state, j)
        loc0 = find_loc(state, 0)
        state = Min
        final.append(state)
    return final
 
#check   
def finish(state, goal):
    return state == goal
```

```
initial = [[2, 8, 3],
           [1, 6, 4],
           [7, 5, 0]]
goal = [[1,2,3],
        [8,0,4],
        [7,6,5]]
print('state: initial')
for i in initial:
  print(i)
print('-----------')
print('state: goal')
for i in goal:
  print(i)
```
