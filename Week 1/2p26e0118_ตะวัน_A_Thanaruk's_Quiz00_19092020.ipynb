0. เขียนโปรแกรมหาค่าสูงสุดในลิสต์

```
import random

#สุ่มเลขจำหนึ่งขึ้นมา
num_lst = [random.randint(10,100) for i in range(20)]
num_lst
```

```
#หาแค่สูงสุด โดยไม่จัดเรียงลำดับ
def maxi_1(lst):
    Max = 0
    for i in lst:
        if i > Max:
            Max = i
    return Max
```

```
maxi_1(num_lst)
```

```
#หาค่าสูงสุดและเรียงลำดับจากมากไปน้อย
def maxi_2(lst):
    for i in range(len(lst)):
        for j in range(len(lst)-1):
            if lst[j] < lst[j+1]:
                lst[j],lst[j+1] = lst[j+1],lst[j]
    print('Max:',lst[0])
    return lst
```

```
maxi_2(num_lst)
```
